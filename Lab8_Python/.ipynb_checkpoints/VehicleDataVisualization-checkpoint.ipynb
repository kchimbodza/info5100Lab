```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
# loading my Dataset:
car_dataframe = pd.read_csv("VehicleDataset/car.csv")
```

```python
car_dataframe.head(5)
```

```python
sbn.histplot(data=car_dataframe, x="transmission", y="fuel")
plt.title("Transmission Types Across Fuel Options")
plt.xlabel("Transmission Type")
plt.ylabel("Fuel Options")
```

```python
# Create a KDE plot
plt.figure(figsize=(8, 6))
sbn.kdeplot(data=car_dataframe, x="year", hue="owner", fill=True, common_norm=False, palette="viridis", alpha=0.2, warn_singular=False)

# Add labels and title
plt.title("Density of Car Onwers by year")
plt.xlabel("Year")
plt.ylabel("Density")
plt.show()
```
